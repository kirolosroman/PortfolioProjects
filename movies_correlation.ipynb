# import libraries

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8)# adjusts the configuration of the plots that will be created

# reading data
df=pd.read_csv(r'D:\Projects\data analysis\movies.csv')
```

```python
df.head()
```

#checking missing data

```python
for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{}-{}%'.format(col,pct_missing))
```

# checking data types

```python
df.dtypes
```

# changing datatype of columns

```python
df=df.dropna()
df = df.astype({"budget":'int64', "gross":'int64'})
df
```

```python
df['yearCorrect']=df['released'].astype(str).str[9:13]
df
```

## sorting

```python
df.sort_values(by=['gross'],inplace=False,ascending=False)
```

# delete any duplicates

```python
df['company'].drop_duplicates().sort_values(ascending=False)
df['company']=df['company'].drop_duplicates()
```

# scatter plot with budget vs gross

```python
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('budget vs gross revenue')
plt.xlabel('budget')
plt.ylabel('gross')

plt.show()
```

# plot budget vs gross using seaborn

```python
sns.regplot(x='budget',y='gross',data=df,scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

# testing correlation

```python
correlation_matrix=df.corr(method='pearson',numeric_only=True)
sns.heatmap(correlation_matrix,annot=True)
plt.title("correlation matrix")
plt.show()
```
